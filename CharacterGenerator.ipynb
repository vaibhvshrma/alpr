```python
def make_textimg(text, font, size=None):
    tx = font.getsize(text)
    img = Image.new(mode="1", size=tx, color=1)
    if size:
        tl = (size[0]-tx[0])//2, (size[1]-tx[1])//2
    else:
        tl = (0,-2)
        draw = ImageDraw.Draw(img)
        draw.text(tl, text, fill='black', font=font)

    return img
```

```python
font = ImageFont.truetype(font="./fonts/Arial.ttf", size=30)
im = make_textimg('A', font)
imshow(np_image(im), cmap="gray")
# im
```

```python
sz = 34     # size of the text
total_fonts = len(os.listdir("./fonts"))
for i,ft in enumerate(os.listdir('./fonts')):
  print(f'Processing font {i+1}/{total_fonts} {ft}')
  font = ImageFont.truetype(f'./fonts/{ft}', sz)
  for j,char in enumerate(chars):
    print(f'{char} ', end='')
    make_textimg(char, font).save(f'./dataset/{char}/{i+1}_{ft.lower()[:ft.rfind(".")]}.jpg')
  print()
```
